```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data = https://aqli.epic.uchicago.edu/the-index/
pm_change = pd.read_csv('./aqli_regional_data_IND.csv')
```

```python
pm_change
```

```python
standard = 10
pm_2000 = pm_change['pm_2000']
pm_drop = pm_2000 - standard
pm_drop = np.clip(pm_drop, a_min=0, a_max=1000000000)
np_pm_drop = np.array(pm_drop)
print(np_pm_drop)
```

```python
life_drop = pm_change['llpp_who_2000']
np_life_drop = np.array(life_drop)
print(np_life_drop)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(np_pm_drop, np_life_drop)
plt.xlabel('AQI value decrease')
plt.ylabel('life expentency increase (years)')
plt.title('life expectancy increase (years) vs AQI decrease')
gradients = np_pm_drop / np_life_drop
gradient = np.median(gradients)

print(f'GRADIENT {gradient} AQI / year')
```
